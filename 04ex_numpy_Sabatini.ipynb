2\. **Outer product**

Find the outer product of the following vectors:

```python
u = np.array([1, 3, 5, 7])
v = np.array([2, 4, 6, 8])
```

Use different methods to do this:

   1. Using the function `outer` in numpy
   2. Using a nested `for` loop or a list comprehension
   3. Using numpy broadcasting operations

````python
# Ex. 2.

u = np.array([1,3,5,7])
v = np.array([2,4,6,8])

print("Outer product using the function outer in numpy:")
print(np.outer(u,v))

print("\nOuter product using a nested for loop or a list comprehension:")
print(np.array([x*y for x in u for y in v]).reshape(4,4))

print("\nOuter product using numpy broadcasting operations:")
zero = np.zeros((len(u),len(v)))
print((zero+v)*((zero+u).T))
````

6\. **Broadcasting**

Use broadcasting to create a grid of distances.

Route 66 crosses the following cities in the US: Chicago, Springfield, Saint-Louis, Tulsa, Oklahoma City, Amarillo, Santa Fe, Albuquerque, Flagstaff, Los Angeles.

The corresponding positions in miles are: 0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448

  * Build a 2D grid of distances among each city along Route 66
  * Convert the distances in km

````python
# Ex. 6.

pos = np.array([0,198,303,736,871,1175,1475,1544,1913,2448])
dist = np.absolute(pos[:,np.newaxis]-pos)

dist_km = (dist * 1.60934)

print("Distances (miles):",'\n',dist,'\n')
print("Distances (km):",'\n',dist_km,'\n')

cities = ["Chicago", "Springfield", "Saint-Louis", "Tulsa", "Oklahoma City", "Amarillo",
          "Santa Fe", "Albuquerque", "Flagstaff", "Los Angeles"]
print("The distance between:")
for i in range(10):
    for k in range(i, 10):
        if i != k:
            print("{} and {} is {} km".format(cities[i], cities[k], round(dist_km[i][k],2)))
````

7\. **Prime numbers sieve**

Compute the prime numbers in the 0-N (start with N=99) range with a sieve (mask).

  * Constract a shape (N,) boolean array, which is the mask
  * Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
  * Apply the mask to obtain an array of ordered prime numbers
  * Check the performances (with `timeit`); how does it scale with N?
  * Implement the optimization suggested in the [sieve of Eratosthenes](https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes)

````python
# Ex. 7.

import numpy as np
import matplotlib.pyplot as plt
import time

def sieve(N):
    #Fills an array of N true boolean variables, which will be the mask
    mask = np.array([True for i in range(N)])
    #Set the first two elements of the mask as false because 0 and 1 are not prime numbers
    mask[0]= False
    if (N>=2):
        mask[1]= False
    #mask[0:2,] = False
    #Performs a for loop from 2 to N-1
    for i in range(2,N):
            #Performs a nested for loop from 2 to i-1
            #It is important that the number i must not be included in the range of the second for loop because a prime number is always divisible by itself
            for j in range(2, i): 
                #If the remainder of i/j is zero
                if (i % j) == 0:
                    #then mask[i] is equal to False
                    mask[i] = False
                    #and exits the second for loop, because if number i is a multiple of another number j then i it is not a prime number
                    break
    #Apply the mask to an array with all numbers from 0 to N-1
    return np.arange(N)[mask]
    #Or as an alternative:
    #return np.array([i for i in range(N) if mask[i]])

print("The prime numbers in the range 0-99 are:")
print(sieve(100))
print("\nWith a sieve (mask) the execution time for prime numbers in the range 0-99 is:")
%timeit sieve(100)

n = []
t = []

for i in range(2,20000,1000):
    start = time.time()
    sieve(i)
    t.append(time.time()-start)
    n.append(i)

plt.clf()
plt.plot(n,t,'o',markersize = 1.5)
plt.xlabel("N maximum of the range [0,N) in which prime numbers are computed")
plt.ylabel("Execution time (s)")
plt.grid()
plt.show()
print("Performance scales roughly exponentially with N")


def sieve_Eratosthenes(N):
    #Fills an array of N true boolean variables, which will be the mask
    mask = [True for i in range(N)]
    #Set the first two elements of the mask as false because 0 and 1 are not prime numbers
    mask[0]= False
    if (N>=2):
        mask[1]= False
    #Scan all numbers p from 2 to N as long as it is worth p*p<N
    p = 2
    while (p*p < N):
        #if the element p is a prime number
        if (mask[p] == True):
            #sets as False in the mask all multiple elements of p
            for i in range(2*p, N, p): 
                mask[i] = False
        p += 1
    #Apply the mask to an array with all numbers from 0 to N-1
    return np.arange(N)[mask]
    #Or as an alternative:
    #return np.array([i for i in range(N) if mask[i]])

print("\nThe prime numbers in the range 0-99 are:")
print(sieve_Eratosthenes(100))
print("\nWith the sieve of Eratostheneshe the execution time for prime numbers in the range 0-99 is:")
%timeit sieve_Eratosthenes(100)
````

8\. **Diffusion using random walk**

Consider a simple random walk process: at each step in time, a walker jumps right or left (+1 or -1) with equal probability. The goal is to find the typical distance from the origin of many random walkers after a given amount of time.

*Hint*: create a 2D array where each row represents a walker, and each column represents a time step.

  * Take 1000 walkers and let them walk for 200 steps
  * Use `randint` to create a 2D array of size $walkers \times steps$ with values -1 or 1
  * Calculate the walking distances for each walker (e.g. by summing the elements in each row)
  * Take the square of the previously-obtained array (element-wise)
  * Compute the mean of the squared distances at each step (i.e. the mean along the columns)
  * **Optional**: plot the average distances ($\sqrt(distance^2)$) as a function of time (step)

````python
# Ex. 8.

import numpy as np
import matplotlib.pyplot as plt

def random_walk(walkers, steps):
    #Creates a 2D array of size walkers x steps with integers random values -1 and 1 
    #Each row of the 2D array represents a walker and each column represents a time step
    random_array =  np.random.choice([-1,+1],(walkers,steps))
    #Calculates the walking distances for each walker by summing the elements in each row
    walking_dist = np.zeros((walkers, steps)) 
    #Scan the rows of the array
    for i in range(np.size(random_array,0)):
        #scan the columns of the array
        for j in range(np.size(random_array,1)):
            #fills each row of walking_dist with the sum of the j elements of each row of random_array
            walking_dist[i,j] = np.sum(random_array[i,:j])  
    return walking_dist

distances = random_walk(1000, 200)
#Computes the squared distances
distances_squared = np.square(distances)
#Computes the mean of the squared distances at each step (i.e. the mean along the columns)
mean_squared = np.mean(distances_squared, axis = 0)
#Computes the square root of the mean squared distances
rms = np.sqrt(mean_squared)

#Plot the root mean square distances as a function of time steps
time_grid = np.linspace(1, 200, 200)
plt.plot(time_grid, rms, label = 'Results')
plt.plot(time_grid, np.sqrt(time_grid), label = 'Prediction: RMS $\propto \sqrt{time}$')
plt.legend()
plt.xlabel('Time steps')
plt.ylabel('RMS distance')
plt.grid()
plt.show()
````
